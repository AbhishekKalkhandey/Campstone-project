### Data Wrangling Code

````python
import pandas as pd
import re

class AmazonWrangling:

    def __init__(self, df):
        self.df = df.copy()

        # Clean runtime safely
        self.df[['runtime_value', 'runtime_unit']] = \
            self.df['runtime'].apply(self._extract_runtime)

    # -------------------------------
    # BASIC GETTERS
    # -------------------------------

    def get_df(self):
        return self.df

    def get_movies(self):
        return self.df[self.df['type'] == 'MOVIE']

    def get_tvshows(self):
        return self.df[self.df['type'] == 'TV SHOW']

    # -------------------------------
    # TOP VALUES
    # -------------------------------

    def get_top_values(self, column, n=10):
        if column not in self.df.columns:
            raise ValueError("Invalid column name")

        return (
            self.df[column]
            .dropna()
            .value_counts()
            .head(n)
        )

    # -------------------------------
    # RUNTIME CLEANING
    # -------------------------------

    def _extract_runtime(self, value):
        if pd.isna(value):
            return pd.Series([None, None])

        value = str(value).lower()

        if "season" in value:
            num = re.findall(r'\d+', value)
            return pd.Series([int(num[0]) if num else None, "season"])
        else:
            num = re.findall(r'\d+', value)
            return pd.Series([int(num[0]) if num else None, "min"])

    # -------------------------------
    # SPLIT MULTI-VALUE COLUMNS
    # -------------------------------

    def get_split_counts(self, column):
        if column not in self.df.columns:
            raise ValueError("Invalid column name")

        temp = (
            self.df[column]
            .dropna()
            .str.split(',')
            .explode()
            .str.strip()
        )

        return temp.value_counts()

    # -------------------------------
    # SHOWS PER MONTH
    # -------------------------------

    def get_shows_per_month(self, typ='all'):

        if typ == 'tv':
            df = self.get_tvshows()
        elif typ == 'movie':
            df = self.get_movies()
        else:
            df = self.df.copy()

        if 'date_added' not in df.columns:
            raise ValueError("date_added column not available")

        df = df.dropna(subset=['date_added'])

        df['added_month'] = pd.to_datetime(df['date_added']).dt.month_name()

        return df['added_month'].value_counts()

    # -------------------------------
    # TEXT LENGTH
    # -------------------------------

    def get_text_length(self, column):
        if column not in self.df.columns:
            raise ValueError("Invalid column name")

        return self.df[column].fillna("").apply(len)
````

````python
amazon = AmazonWrangling(df)
````
